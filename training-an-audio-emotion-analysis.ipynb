```python
import os
import numpy as np
import pandas as pd
import librosa
import librosa.display
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models, utils, backend as K
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import warnings
import joblib # For saving scaler and encoder

warnings.filterwarnings('ignore')

# --- Configuration ---
# Using Kaggle's standard input directory structure
DATASET_PATH = "/kaggle/input/ravdess-emotional-speech-audio"
# Consider saving processed features to Kaggle's working directory for faster re-runs
# PROCESSED_FEATURES_PATH = "/kaggle/working/ravdess_features.pkl"
# PROCESSED_LABELS_PATH = "/kaggle/working/ravdess_labels.pkl"
MODEL_SAVE_PATH = "/kaggle/working/best_emotion_model.keras"
SCALER_SAVE_PATH = "/kaggle/working/scaler.joblib"
ENCODER_SAVE_PATH = "/kaggle/working/label_encoder.joblib"

# Audio Parameters
SAMPLE_RATE = 22050
DURATION = 3  # Max duration to load/process
N_MFCC = 40 # Number of MFCCs to extract

# Training Parameters
BATCH_SIZE = 32
EPOCHS = 150 # Increased epochs, but EarlyStopping will prevent overfitting
VALIDATION_SPLIT = 0.2
RANDOM_STATE = 42

# Emotion mapping based on RAVDESS filename convention
EMOTION_MAP = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

# --- Feature Extraction Functions ---

def parse_ravdess_filename(filename: str) -> str | None:
    """
    Parses the RAVDESS filename to extract the emotion code.
    Example: 03-01-03-01-01-01-12.wav -> '03' -> 'happy'
    Returns the emotion string or None if parsing fails.
    """
    try:
        parts = filename.split('.')[0].split('-')
        emotion_code = parts[2]
        return EMOTION_MAP[emotion_code]
    except (IndexError, KeyError):
        print(f"Warning: Could not parse emotion from filename: {filename}")
        return None

def extract_detailed_features(audio: np.ndarray, sr: int) -> np.ndarray | None:
    """
    Extracts MFCCs, pitch, intensity (RMS energy), and speech rate features from an audio array.

    Args:
        audio (np.ndarray): The audio time series.
        sr (int): The sample rate of the audio.

    Returns:
        np.ndarray: A flat numpy array containing the extracted features,
                    or None if an error occurs (e.g., audio too short for pitch).
                    The feature order is: [mean_mfccs, pitch_mean, pitch_std, intensity, speech_rate]
    """
    try:
        # MFCCs - Captures spectral envelope
        # Using hop_length=512 for better time resolution if needed, adjust if necessary
        mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=N_MFCC, n_fft=2048, hop_length=512)
        mfccs_mean = np.mean(mfccs, axis=1)

        # Pitch (Fundamental Frequency - F0) using YIN algorithm
        # yin requires a minimum duration; short clips might cause issues
        f0, voiced_flag, voiced_probs = librosa.pyin(audio, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'), sr=sr)
        # Replace NaNs (unvoiced frames) with 0 or handle differently if needed
        # Here, we only consider voiced frames for mean/std calculation
        pitch_voiced = f0[voiced_flag]
        if len(pitch_voiced) > 0:
            pitch_mean = np.mean(pitch_voiced)
            pitch_std = np.std(pitch_voiced)
        else:
            # Handle cases with no detected pitch (e.g., very short clips, silence, noise)
            pitch_mean = 0.0
            pitch_std = 0.0

        # Intensity (Root Mean Square Energy) - Correlates with loudness
        rms = librosa.feature.rms(y=audio, frame_length=2048, hop_length=512)
        intensity = np.mean(rms)

        # Speech Rate (simple estimation based on silences)
        # Threshold might need adjustment based on recording quality
        non_silent_intervals = librosa.effects.split(audio, top_db=30) # Use a reasonable top_db
        if len(audio) > 0 and len(non_silent_intervals) > 0:
             # Calculate total duration of speech segments
             speech_duration = sum(interval[1] - interval[0] for interval in non_silent_intervals) / sr
             total_duration = len(audio) / sr
             # Avoid division by zero if total_duration is somehow zero
             speech_rate = len(non_silent_intervals) / total_duration if total_duration > 0 else 0
        else:
             speech_rate = 0 # No speech detected or empty audio

        # --- Combine features ---
        # Order: N_MFCC means, pitch_mean, pitch_std, intensity, speech_rate
        feature_vector = np.hstack((mfccs_mean, pitch_mean, pitch_std, intensity, speech_rate))

        # Ensure no NaNs or Infs made it through (should be handled above, but belt-and-suspenders)
        if np.any(np.isnan(feature_vector)) or np.any(np.isinf(feature_vector)):
            print(f"Warning: NaN or Inf found in feature vector for a segment. Replacing with 0.")
            feature_vector = np.nan_to_num(feature_vector, nan=0.0, posinf=0.0, neginf=0.0)

        return feature_vector

    except Exception as e:
        # More specific error logging
        print(f"Error extracting features: {e}")
        # You could add filename here if passed down
        return None


def load_and_extract_features(file_path: str, sr: int = SAMPLE_RATE, duration: float = DURATION) -> np.ndarray | None:
    """Loads an audio file and extracts features using extract_detailed_features."""
    try:
        # Load audio file, limiting duration
        audio, current_sr = librosa.load(file_path, sr=sr, duration=duration, res_type='kaiser_fast')

        # Pad or truncate if necessary to ensure consistent length (optional, depends on model needs)
        # target_length = int(sr * duration)
        # if len(audio) < target_length:
        #     audio = np.pad(audio, (0, target_length - len(audio)), mode='constant')
        # elif len(audio) > target_length:
        #     audio = audio[:target_length]

        if len(audio) == 0:
             print(f"Warning: Loaded audio is empty for file {file_path}. Skipping.")
             return None

        # Extract features using the detailed function
        return extract_detailed_features(audio, sr)

    except Exception as e:
        print(f"Error loading or processing {file_path}: {e}")
        return None

# --- Data Preparation ---

def prepare_dataset(dataset_path: str):
    """
    Walks through the dataset path, loads audio files, extracts features,
    and collects corresponding emotion labels.

    Args:
        dataset_path (str): The root path of the RAVDESS dataset.

    Returns:
        tuple[np.ndarray, np.ndarray]: A tuple containing:
            - features (np.ndarray): An array of extracted feature vectors.
            - labels (np.ndarray): An array of corresponding emotion labels.
    """
    all_features = []
    all_labels = []
    print(f"Starting dataset preparation from: {dataset_path}")
    file_count = 0
    processed_count = 0
    error_count = 0

    for root, dirs, files in os.walk(dataset_path):
        # Limit processing to actor folders if structure is known (e.g., Actor_01, Actor_02...)
        # This avoids potentially processing other files in the root dataset_path
        # basedir = os.path.basename(root)
        # if not basedir.startswith("Actor_"): continue

        for filename in files:
            if filename.endswith('.wav'):
                file_count += 1
                file_path = os.path.join(root, filename)
                emotion = parse_ravdess_filename(filename)

                if emotion:
                    feature_vector = load_and_extract_features(file_path, sr=SAMPLE_RATE, duration=DURATION)

                    if feature_vector is not None:
                        all_features.append(feature_vector)
                        all_labels.append(emotion)
                        processed_count += 1
                    else:
                        error_count += 1
                else:
                    error_count += 1 # Count parsing errors as well

                if file_count % 200 == 0: # Print progress update
                    print(f"  Processed {file_count} files...")

    print(f"\nDataset Preparation Summary:")
    print(f"  Total .wav files found: {file_count}")
    print(f"  Successfully processed: {processed_count}")
    print(f"  Skipped due to errors: {error_count}")

    if not all_features:
        print("Error: No features were extracted. Check dataset path, file formats, or feature extraction logic.")
        return np.array([]), np.array([]) # Return empty arrays

    return np.array(all_features), np.array(all_labels)

# --- Model Definition ---

def create_emotion_model(input_shape: tuple, num_classes: int) -> models.Model:
    """
    Creates a Sequential Deep Neural Network model for emotion classification.
    """
    model = models.Sequential([
        layers.Input(shape=input_shape, name='input_features'),

        # Batch Normalization helps stabilize training and can act as regularization
        layers.BatchNormalization(),

        # Dense layers with ReLU activation and Dropout for regularization
        layers.Dense(256, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        layers.Dropout(0.4), # Increased dropout slightly

        layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        layers.Dropout(0.4),

        layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
        layers.Dropout(0.3),

        # Output layer with Softmax activation for multi-class classification
        layers.Dense(num_classes, activation='softmax', name='output_emotion')
    ])

    # Using Adam optimizer with a potentially lower initial learning rate
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy', # Standard loss for multi-class classification
        metrics=['accuracy']
    )

    return model

# --- Rule-Based Refinement (Example for Happy/Angry) ---

def is_likely_happy(scaled_features_single_sample: np.ndarray) -> bool:
    """
    Applies heuristic rules based on scaled acoustic features to check
    if an audio sample characteristics align with 'happy'.
    NOTE: This relies heavily on the feature scaling and the specific dataset.
          Thresholds likely need tuning based on analysis of scaled feature distributions.

    Args:
        scaled_features_single_sample (np.ndarray): A 1D array of SCALED features
            for a single sample. Assumes the standard feature order:
            [... MFCCs ..., pitch_mean, pitch_std, intensity, speech_rate]

    Returns:
        bool: True if the features match 'happy' characteristics, False otherwise.
    """
    # Indices assuming the feature order from extract_detailed_features:
    # N_MFCC means, then pitch_mean, pitch_std, intensity, speech_rate
    PITCH_MEAN_IDX = N_MFCC
    PITCH_STD_IDX = N_MFCC + 1
    INTENSITY_IDX = N_MFCC + 2
    # SPEECH_RATE_IDX = N_MFCC + 3 # Not used in this rule, but index is known

    # Ensure the input is a 1D array
    if scaled_features_single_sample.ndim > 1:
       scaled_features_single_sample = scaled_features_single_sample.flatten()

    # Check if indices are valid
    if len(scaled_features_single_sample) <= max(PITCH_MEAN_IDX, PITCH_STD_IDX, INTENSITY_IDX):
        print("Warning: Scaled feature vector too short for rule-based check.")
        return False

    pitch_mean_scaled = scaled_features_single_sample[PITCH_MEAN_IDX]
    pitch_std_scaled = scaled_features_single_sample[PITCH_STD_IDX]
    intensity_scaled = scaled_features_single_sample[INTENSITY_IDX]

    # --- Define Thresholds (These MUST be tuned based on SCALED data analysis!) ---
    # Example thresholds - Replace with values derived from your scaled data analysis
    # 'Happy' often has higher avg pitch, moderate pitch variation, moderate-to-high intensity.
    # The exact scaled values depend heavily on the overall dataset distribution.
    # Let's assume StandardScaler results in roughly mean=0, std=1.
    # Higher than average pitch mean (e.g., > 0 standard deviations)
    pitch_mean_threshold = 0.0
    # Low-to-moderate pitch variation (e.g., < 1 standard deviation)
    pitch_std_threshold = 1.0
    # Moderate-to-high intensity (e.g., > -0.5 standard deviations)
    intensity_threshold_low = -0.5
    intensity_threshold_high = 1.5 # Upper bound might also be useful

    # --- Apply Rules ---
    is_happy_pitch_mean = pitch_mean_scaled > pitch_mean_threshold
    is_stable_pitch = pitch_std_scaled < pitch_std_threshold
    is_moderate_intensity = intensity_threshold_low < intensity_scaled < intensity_threshold_high

    # Combine rules (adjust logic as needed)
    # Here: Requires happy pitch AND stable pitch AND moderate intensity
    return is_happy_pitch_mean and is_stable_pitch and is_moderate_intensity

# --- Augmentation (for Prediction/TTA) ---

def augment_audio(audio: np.ndarray, sr: int) -> np.ndarray:
    """
    Applies a randomly selected augmentation suitable for emotion analysis.
    """
    choice = np.random.choice(['pitch', 'speed', 'noise', 'none'])

    if choice == 'pitch':
        # Subtle pitch shifts
        steps = np.random.uniform(-1.5, 1.5) # Slightly wider range than before
        return librosa.effects.pitch_shift(y=audio, sr=sr, n_steps=steps)
    elif choice == 'speed':
        # Subtle speed changes
        rate = np.random.uniform(0.85, 1.15) # Slightly wider range
        # Stretch audio without changing pitch
        return librosa.effects.time_stretch(y=audio, rate=rate)
    elif choice == 'noise':
        # Add low-level Gaussian noise
        noise_amp = 0.003 * np.random.uniform(0.5, 1.2) * np.max(np.abs(audio))
        noise = np.random.normal(0, noise_amp, len(audio))
        return audio + noise
    else: # choice == 'none'
        return audio

# --- Prediction Function ---

def predict_emotion_with_tta_and_rules(
    file_path: str,
    model: models.Model,
    label_encoder: LabelEncoder,
    scaler: StandardScaler,
    n_augments: int = 5,
    rule_based_refinement: bool = True
) -> tuple[str | None, float]:
    """
    Predicts emotion for a single audio file using the trained model,
    optionally applying Test-Time Augmentation (TTA) and rule-based refinement.

    Args:
        file_path (str): Path to the audio file.
        model (models.Model): The trained Keras model.
        label_encoder (LabelEncoder): The fitted LabelEncoder instance.
        scaler (StandardScaler): The fitted StandardScaler instance.
        n_augments (int): Number of augmented versions to create for TTA (0 disables TTA).
        rule_based_refinement (bool): Whether to apply the 'is_likely_happy' rule.

    Returns:
        tuple[str | None, float]: A tuple containing:
            - Predicted emotion label (str) or None if error.
            - Confidence score (float) for the prediction.
    """
    # 1. Extract features from the original audio
    main_features = load_and_extract_features(file_path, sr=SAMPLE_RATE, duration=DURATION)
    if main_features is None:
        print(f"Could not extract features for prediction from: {file_path}")
        return None, 0.0

    # 2. Scale the main features
    main_features_scaled = scaler.transform(main_features.reshape(1, -1))

    # 3. Get initial prediction from the original audio
    base_pred_probs = model.predict(main_features_scaled, verbose=0)[0] # Get probability vector

    # Store predictions (probabilities)
    all_pred_probs = [base_pred_probs]

    # 4. Test-Time Augmentation (TTA)
    if n_augments > 0:
        try:
            audio, sr = librosa.load(file_path, sr=SAMPLE_RATE, duration=DURATION)
            if len(audio) > 0:
                for _ in range(n_augments):
                    augmented_audio = augment_audio(audio, sr)
                    aug_features = extract_detailed_features(augmented_audio, sr)
                    if aug_features is not None:
                        aug_features_scaled = scaler.transform(aug_features.reshape(1, -1))
                        aug_pred_probs = model.predict(aug_features_scaled, verbose=0)[0]
                        all_pred_probs.append(aug_pred_probs)
            else:
                print(f"Warning: Cannot perform TTA, audio empty for {file_path}")

        except Exception as e:
            print(f"Error during TTA for {file_path}: {e}")
            # Proceed with base prediction if TTA fails

    # 5. Combine Predictions (Average Probabilities)
    if not all_pred_probs: # Should not happen if main extraction worked, but check
         print(f"Error: No predictions available for {file_path}")
         return None, 0.0

    avg_pred_probs = np.mean(all_pred_probs, axis=0)
    pred_idx = np.argmax(avg_pred_probs)
    predicted_emotion = label_encoder.inverse_transform([pred_idx])[0]
    confidence = float(np.max(avg_pred_probs)) # Ensure it's a standard float

    # 6. Rule-Based Refinement (Optional)
    if rule_based_refinement:
        # Example: Refine high-confidence 'angry' predictions that look 'happy' acoustically
        angry_idx = label_encoder.transform(['angry'])[0]
        happy_idx = label_encoder.transform(['happy'])[0]

        # Check if the highest probability is 'angry' with high confidence
        if pred_idx == angry_idx and confidence > 0.80:
             # Pass the SCALED features of the ORIGINAL audio to the rule function
            if is_likely_happy(main_features_scaled):
                print(f"  Rule applied: Overriding high-confidence 'angry' to 'happy' for {os.path.basename(file_path)}")
                # Return 'happy', but maybe slightly reduce confidence to indicate override
                predicted_emotion = 'happy'
                confidence = min(confidence, 0.85) # Cap confidence slightly

        # --- Add more rules here if needed (e.g., for sad/calm confusion) ---

    return predicted_emotion, confidence


# --- Main Execution ---

if __name__ == "__main__": # Ensures code runs only when script is executed directly

    # === 1. Data Loading and Feature Extraction ===
    # Optional: Check if pre-processed features exist
    # try:
    #     print("Attempting to load pre-processed features...")
    #     X = joblib.load(PROCESSED_FEATURES_PATH)
    #     y_labels = joblib.load(PROCESSED_LABELS_PATH)
    #     print("Loaded pre-processed features successfully.")
    # except FileNotFoundError:
    #     print("Pre-processed features not found. Extracting from audio files...")
    X, y_labels = prepare_dataset(DATASET_PATH)
        # if X.shape[0] > 0:
        #     print("Saving processed features for future runs...")
        #     joblib.dump(X, PROCESSED_FEATURES_PATH)
        #     joblib.dump(y_labels, PROCESSED_LABELS_PATH)

    # Exit if data loading failed
    if X.shape[0] == 0 or y_labels.shape[0] == 0:
        print("Fatal: Dataset loading/feature extraction failed. Cannot proceed.")
        exit()

    print(f"\nLoaded {X.shape[0]} samples with {X.shape[1]} features each.")

    # === 2. Data Preprocessing ===
    print("\nPreprocessing data...")
    # Encode labels
    label_encoder = LabelEncoder()
    y_encoded = label_encoder.fit_transform(y_labels)
    num_classes = len(label_encoder.classes_)
    print(f"Found {num_classes} emotion classes: {label_encoder.classes_}")

    # Convert labels to one-hot encoding for categorical_crossentropy
    y_one_hot = utils.to_categorical(y_encoded, num_classes=num_classes)

    # Split data (stratified to maintain class distribution)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_one_hot,
        test_size=VALIDATION_SPLIT,
        random_state=RANDOM_STATE,
        stratify=y_encoded # Use original encoded labels for stratification
    )
    print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")

    # Scale features (Fit ONLY on training data, transform both train and test)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Save the scaler and encoder for later use during prediction
    print("Saving scaler and label encoder...")
    joblib.dump(scaler, SCALER_SAVE_PATH)
    joblib.dump(label_encoder, ENCODER_SAVE_PATH)
    print(f"Scaler saved to: {SCALER_SAVE_PATH}")
    print(f"Encoder saved to: {ENCODER_SAVE_PATH}")


    # === 3. Model Creation and Training ===
    print("\nCreating and training the model...")
    input_shape = (X_train.shape[1],)
    model = create_emotion_model(input_shape, num_classes)
    model.summary()

    # Callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',      # Monitor validation loss
        patience=20,             # Increase patience slightly
        verbose=1,
        restore_best_weights=True # Restore weights from the epoch with the best val_loss
    )
    model_checkpoint = ModelCheckpoint(
        MODEL_SAVE_PATH,         # Save path in Kaggle working directory
        monitor='val_accuracy',  # Monitor validation accuracy
        save_best_only=True,     # Only save the best model
        verbose=1
    )
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',      # Monitor validation loss
        factor=0.2,              # Reduce LR by a factor of 5
        patience=7,              # Reduce LR if no improvement after 7 epochs
        min_lr=0.00001,          # Minimum learning rate
        verbose=1
    )

    # Train the model
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        callbacks=[early_stopping, model_checkpoint, reduce_lr],
        verbose=1 # Set to 1 or 2 for progress bars in Kaggle
    )

    # === 4. Evaluation ===
    print("\nEvaluating the best model on the test set...")
    # The best weights are either restored by EarlyStopping or loaded via ModelCheckpoint
    # If EarlyStopping restored weights, loading is redundant but safe.
    # If training finished fully, ModelCheckpoint ensures we load the best saved version.
    try:
        print(f"Loading best model from {MODEL_SAVE_PATH}")
        model = models.load_model(MODEL_SAVE_PATH)
    except Exception as e:
         print(f"Could not load saved model: {e}. Using the model from the end of training.")
         # Note: If early stopping restored weights, this model is already the best one.

    test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
    print(f"Test Loss: {test_loss:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")

    # Plot training history (Loss and Accuracy)
    print("\nPlotting training history...")
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')

    plt.tight_layout()
    plt.savefig('/kaggle/working/training_history.png') # Save plot in Kaggle output
    plt.show()


    # Detailed evaluation: Classification Report and Confusion Matrix
    print("\nGenerating Classification Report and Confusion Matrix...")
    y_pred_probs = model.predict(X_test)
    y_pred_labels = np.argmax(y_pred_probs, axis=1)
    y_true_labels = np.argmax(y_test, axis=1) # Convert one-hot back to integer labels

    print("\nClassification Report:")
    print(classification_report(y_true_labels, y_pred_labels, target_names=label_encoder.classes_))

    print("\nConfusion Matrix:")
    cm = confusion_matrix(y_true_labels, y_pred_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.savefig('/kaggle/working/confusion_matrix.png') # Save plot
    plt.show()


    # === 5. Example Prediction ===
    print("\nRunning prediction on a few test samples...")

    # Load the scaler and encoder we saved earlier
    try:
        loaded_scaler = joblib.load(SCALER_SAVE_PATH)
        loaded_encoder = joblib.load(ENCODER_SAVE_PATH)
        print("Loaded scaler and encoder for prediction.")

        # Find a few example files from the original dataset path to test prediction
        # Note: These examples were likely part of train or test set,
        # ideally predict on *new* unseen data if available.
        example_files = []
        files_needed = 3
        for root, dirs, files in os.walk(DATASET_PATH):
             for filename in files:
                  if filename.endswith('.wav') and len(example_files) < files_needed:
                      example_files.append(os.path.join(root, filename))
             if len(example_files) >= files_needed:
                 break

        if not example_files:
             print("Could not find example files in the dataset path for prediction.")
        else:
             print("\n--- Example Predictions ---")
             for file_path in example_files:
                 true_emotion = parse_ravdess_filename(os.path.basename(file_path))
                 predicted_emotion, confidence = predict_emotion_with_tta_and_rules(
                     file_path,
                     model,
                     loaded_encoder,
                     loaded_scaler,
                     n_augments=5, # Use TTA
                     rule_based_refinement=True # Use rule refinement
                 )
                 print(f"File: {os.path.basename(file_path)}")
                 print(f"  True Emotion: {true_emotion}")
                 print(f"  Predicted Emotion: {predicted_emotion} (Confidence: {confidence:.3f})")
                 print("-" * 20)

    except FileNotFoundError:
        print("Error: Could not load saved scaler or encoder. Cannot run prediction example.")
    except Exception as e:
        print(f"An error occurred during example prediction: {e}")

    print("\n--- Script Finished ---")
```
